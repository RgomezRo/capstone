```python
class TimeSeriesFAO:


  __arrayOfFeatures__ = []
  
  
  def __init__(self,df,categoryToMeasure,categoryLabel):
    self.df = df[df["Element"]== categoryToMeasure ]
    self.categoryToMeasure = categoryToMeasure
    self.categoryLabel = categoryLabel

  def viewDF(self,numberOfRows=1):
    print("SAMPLE DATA:\n")
    display(self.df.head(numberOfRows))
  
  def prepareData(self):
    
    # deal with null values 
    
    print("\nnull values before:\r\n")
    display(pd.DataFrame(self.df.isnull().sum()).T)

    self.df["Value"].fillna(self.df["Value"].mean(), inplace=True)
    
    print("\nnull values after:\r\n")
    display(pd.DataFrame(self.df.isnull().sum()).T)

    # delete constant values

    del(self.df["Flag"])

    # delete aggregate country rows
    
    indexAreas = pd.DataFrame(data=['World',
  'Africa',
  'Eastern Africa',
  'Middle Africa',
  'Northern Africa',
  'Southern Africa',
  'Western Africa',
  'Americas',
  'Northern America',
  'Central America',
  'Caribbean',
  'South America',
  'Asia',
  'Central Asia',
  'Eastern Asia',
  'Southern Asia',
  'South-eastern Asia',
  'Western Asia',
  'Europe',
  'Eastern Europe',
  'Northern Europe',
  'Southern Europe',
  'Western Europe',
  'Oceania',
  'Australia and New Zealand',
  'Melanesia',
  'Micronesia',
  'Polynesia',
  'European Union (27)',
  'Least Developed Countries',
  'Land Locked Developing Countries',
  'Small Island Developing States',
  'Low Income Food Deficit Countries',
  'Net Food Importing Developing Countries'])

    indexAreas["isArea"] = 1
    
    
    
    indexAreas.rename(columns={0: "Area"},inplace=True)
    print("\nAGGREGATIONS OF COUNTRIES TO DROP (sample): \n\n")
  
    display( indexAreas.head(7))
    self.df=self.df.merge(indexAreas, how='left', on='Area')
    
    print("\nShape with aggregated areas : \n\n",self.df.shape)
  
    indexNames = self.df[self.df['isArea']==1].index
    
    self.df.drop(indexNames,inplace=True)
    self.df.reset_index()

    self.df.drop('isArea', inplace=True, axis=1)
    
    print("\nShape without aggregated areas : \n\n",self.df.shape)
  
  def globalRepresentation(self):

    color = ["#ffa69e","#faf3dd","#b8f2e6","#aed9e0","#5e6472",'#f6bd60','#84a59d','#f8edeb']
    customPalette = sns.set_palette(sns.color_palette(color))

    itemMaps = self.df.groupby('Item')['Value'].sum().reset_index()
    group = itemMaps[itemMaps["Value"]>(itemMaps["Value"].sum()*(3.5/100))]
    
    measurement =  self.df["Unit"][0]
    #Como tenemos 115 categor√≠as solo pintamos aquellas que representan mas del 5 % de los valores
    
    
    fig = plt.figure(figsize=(30,30))
    ax1 = fig.add_subplot(311)
    _ = sns.barplot(data=group, y='Item', x='Value', palette= customPalette, ax=ax1)
    ylabels = group['Item'].to_list()
    xlabels = group['Value']
    _ = ax1.set_title(self.categoryLabel + ' per Item', fontsize=25)
    _ = ax1.set_xlabel("Quantity( "+measurement+" )", fontsize=20)
    _ = ax1.set_ylabel('')
    _ = ax1.set_yticklabels(ylabels, rotation=0, fontsize=20)

    fig.tight_layout(pad=0.5)

    fig.show()
    

  def specificRepresentationByItem(self,itemName):
    print(self.df.head(10))

  def specificRepresentationByCountry(self,itemName):
    print(self.df.head(10))
  
  def ma_analysis(self,arrayOfMAs, period):
    print(self.df.head(10))
  
  def variation_analisis(self):
    print(self.df.head(10))
  
  def correlationCalculation(self):
    print(self.df.head(10))
  
  def decompose(self):
    print(self.df.head(10))
  
  def generateAdditionalsFeatures(self,groupByArray, generateMAs,operation,p,q,targetToSummarize):
    print(self.df.head(10))
  
  def mergeAdditionalFeatures(self):
    print(self.df.head(10))
  
  def dropColumns(self):
    print(self.df.head(10))
  
  def geneerateEnvironments(self):
    print(self.df.head(10))
  
  def train(self):
    print(self.df.head(10))

  def featureImportance(self):
    print(self.df.head(10))

  def predictionAndModelEvaluation(self):
    print(self.df.head(10))




    
```
